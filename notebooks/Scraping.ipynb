# Scraping ubicacion

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import time

# Lista de URLs con reviews de Google Maps
urls = df.drop_duplicates('name')[['place_id', 'name', 'link_review']]#.iloc[:20]

# Configuración de Selenium con WebDriver Manager
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Opcional: Ejecuta en modo sin interfaz gráfica
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

# Inicializa el navegador
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Función para extraer los datos de cada URL
def scrape_reviews(url):
    #print(f"Procesando: {url}")

    dict_result = dict()
    dict_result['URL'] = url

    try:
        driver.get(url)
        time.sleep(1)  # Espera unos segundos para cargar la página
        # Esperar a que la URL cambie dinámicamente
        WebDriverWait(driver, 15).until(
            lambda driver: "@-" in driver.current_url
        )
        # Obtener la URL actualizada
        dynamic_url = driver.current_url
        # Buscar las coordenadas en la URL actualizada
        match = re.search(r'@(-?\d+\.\d+),(-?\d+\.\d+)', dynamic_url)
        if match:
            latitude = float(match.group(1))
            longitude = float(match.group(2))
            dict_result['latitud'] = latitude
            dict_result['longitud'] = longitude
        else:
            dict_result['latitud'] = ""
            dict_result['longitud'] = ""

        # Extraer categorias
        parents_span = driver.find_elements(By.XPATH, "//div[@class='RDQN4b']")
        dict_result['categoria'] = parents_span[0].text

    except:
        dict_result['latitud'] = ""
        dict_result['longitud'] = ""
        dict_result['categoria'] = ""

    return dict_result


# Iterar sobre todas las URLs y recolectar datos
all_reviews = []
urls_len= len(urls.link_review.tolist())
i=0
for url in urls.link_review.tolist():
    print(f"{int(100*i/(urls_len-1))}%", end="\r")
    review_data = scrape_reviews(url)
    all_reviews.append(review_data)
    i+=1

# Cierra el navegador
driver.quit()

print()
all_reviews_df = pd.DataFrame(all_reviews)

urls = urls.merge(all_reviews_df, left_on='link_review', right_on='URL')
urls.categoria = urls.categoria.apply(lambda x: x.split(" · ")[0]) # limpia la categoria restaurante, elimina el precio
urls.to_csv('Ubicación_place_id.csv', index=False)
urls.head(20) # demora ~2hrs
```
